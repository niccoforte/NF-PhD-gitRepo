```python
import numpy as np
import pandas as pd
import os
```

```python
inpFile = 'tensile-disNodes/FCC_12X16_per.inp'
```

```python
def get_nodes(inpFile):
    with open(inpFile, 'r') as f:
        lines = f.readlines()
        
    nodes_start = int([lines.index(line) for line in lines if "Node" in line][0]) + 1
    nodes_end = int([lines.index(line) for line in lines if "Element" in line][0])
    
    node_lines = lines[nodes_start:nodes_end]
    node_lines = [line.split(',') for line in node_lines]
    
    perNodes = [[float(elem.strip('\n').strip('.').strip()) for elem in line] for line in node_lines]
    perNodesCoords = [node[1:] for node in perNodes]
    nodes_df = pd.DataFrame(perNodesCoords, columns=['x','y'])
    
    return perNodes, perNodesCoords, nodes_df

def set_boundaries(nodes_df):
    x0 = min(nodes_df.x.tolist())
    x1 = max(nodes_df.x.tolist())
    y0 = min(nodes_df.y.tolist())
    y1 = max(nodes_df.y.tolist())
    
    left = nodes_df[nodes_df.x == x0]
    right = nodes_df[nodes_df.x == x1]
    bot = nodes_df[nodes_df.y == y0]
    top = nodes_df[nodes_df.y == y1]
    
    boundary = left.append(right).append(bot).append(top)
    
    return boundary
```

```python
perNodes, perNodesCoords, nodes_df = get_nodes(inpFile=inpFile)
```

```python
boundary = set_boundaries(nodes_df=nodes_df)
```
